# Coupled RLC circuits

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve, exp, I as j

# Define circuit parameters
V01_val = 48  # Voltage amplitude for first circuit (V)
V02_val = 48  # Voltage amplitude for second circuit (V)
R1_val, R2_val = 0.26, 0.26  # Resistances (ohms)
L1_val, L2_val = 44.45e-6, 44.45e-6  # Inductances (H)
C1_val, C2_val = 468e-9, 468e-9  # Capacitances (F)
k = 0.001  # Coupling coefficient
M_val = k * (L1_val * L2_val)**(1/2)  # Mutual inductance (H)
fm = 100  # Modulation frequency (Hz)
fc = 34.63e3  # Carrier frequency (Hz)
omega_val = 2 * np.pi * fc  # Carrier angular frequency (rad/s)
omega_mod = 2 * np.pi * fm  # Modulation angular frequency (rad/s)

# Define symbolic variables
V01, V02, Z1, Z2, I1, I2 = symbols('V01 V02 Z1 Z2 I1 I2', complex=True)
omega, M, phi, L1, L2, R1, R2 = symbols('omega M phi L1 L2 R1 R2', real=True)

# Define the equations based on the phasor domain analysis (including mutual inductance)
eq1 = Eq(V01, Z1*I1 + j*omega*M*I2)
eq2 = Eq(V02 * exp(j*phi), Z2*I2 + j*omega*M*I1)

# Solve the equations for I1 and I2
solutions = solve([eq1, eq2], (I1, I2))
I1_solution = solutions[I1]
I2_solution = solutions[I2]

print('I1 =', I1_solution)
print('I2 =', I2_solution)
```

## 1. Current Magnitude vs Phase

```python

# Calculate numeric impedances
Z1_val = R1_val + j*omega_val*L1_val - j/(omega_val*C1_val)
Z2_val = R2_val + j*omega_val*L2_val - j/(omega_val*C2_val)

phi_values = np.arange(0, 2*np.pi, 0.01)
I_1_magnitude = []
I_2_magnitude = []


for phi_val in phi_values:
# Substitute the numeric values into the solutions
    I1_numeric = I1_solution.subs({V01: V01_val, V02: V02_val, Z1: Z1_val, Z2: Z2_val, M: M_val, omega: omega_val, phi: phi_val})
    I2_numeric = I2_solution.subs({V01: V01_val, V02: V02_val, Z1: Z1_val, Z2: Z2_val, M: M_val, omega: omega_val, phi: phi_val})

    # Evaluate the solutions numerically
    I1_numeric_eval = I1_numeric.evalf()
    I2_numeric_eval = I2_numeric.evalf()

    print('I1 (numeric) =', I1_numeric_eval)
    print('I2 (numeric) =', I2_numeric_eval)
    I_1_magnitude.append(np.abs(I1_numeric_eval))
    I_2_magnitude.append(np.abs(I2_numeric_eval))

phi_values_pi = phi_values / np.pi
plt.figure(figsize=(10, 6))
plt.plot(phi_values_pi, I_1_magnitude, label='|I1|')
plt.plot(phi_values_pi, I_2_magnitude, label='|I2|')
plt.xlabel(f'Phase difference ($\Phi$) $\pi$')
plt.ylabel('Current magnitude [A]')
plt.title('Current Magnitudes vs Phase Difference')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Amplitude and Phase vs. (angular) frequency

```python
# Define a range of angular frequency values

omega_c = 2 * np.pi * fc
omega_values = np.linspace(np.pi * fc, 4 * np.pi * fc, 1000)
omega_values_normalized = omega_values / omega_c  # Normalize by omega_c


for phi_val in [0, np.pi/2, np.pi, 3*np.pi/2]:
    I1_omega_magnitude = []
    I2_omega_magnitude = []
    # Loop over omega values
    for omega_val in omega_values:
        # Update the impedances for each frequency
        Z1_val = R1_val + j*omega_val*L1_val - j/(omega_val*C1_val)
        Z2_val = R2_val + j*omega_val*L2_val - j/(omega_val*C2_val)
        
        # Substitute the numeric values into the solutions
        I1_numeric = I1_solution.subs({V01: V01_val, V02: V02_val, Z1: Z1_val, Z2: Z2_val, M: M_val, omega: omega_val, phi: phi_val})
        I2_numeric = I2_solution.subs({V01: V01_val, V02: V02_val, Z1: Z1_val, Z2: Z2_val, M: M_val, omega: omega_val, phi: phi_val})

        # Evaluate the solutions numerically
        I1_numeric_eval = I1_numeric.evalf()
        I2_numeric_eval = I2_numeric.evalf()

        # Append magnitudes of currents to the lists
        I1_omega_magnitude.append(np.abs(I1_numeric_eval))
        I2_omega_magnitude.append(np.abs(I2_numeric_eval))

    phi_value = phi_val / np.pi
    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(omega_values_normalized, I1_omega_magnitude, label='|I1|')
    plt.plot(omega_values_normalized, I2_omega_magnitude, label='|I2|')
    plt.xlabel('Angular frequency ($\omega / \omega_c$)')
    plt.ylabel('Current magnitude [A]')
    plt.title(f'Current Magnitudes vs Angular Frequency for $\Phi$ = {phi_value} $\pi$')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def square_wave_delay(ts, fc):
    Vs = []
    
    period = 1/fc  # Full period of the square wave with high, zero, low, zero
    D = period*0.3  # Duration of high voltage
    d = period*0.2  # Duration of zero voltage after high voltage
    for t in ts:
        # Find the time within the current period
        t_mod = t % period
        
        if t_mod < D:
            # High voltage
            V = 48
        elif D <= t_mod < D + d:
            # Zero voltage after high voltage
            V = 0
        elif D + d <= t_mod < 2 * D + d:
            # Negative voltage after zero voltage
            V = -48
        else:
            # Zero voltage before switching back to high
            V = 0
        
        Vs.append(V)
    
    return Vs
```

```python
# fourier transform of the square wave

from scipy.fft import fft, ifft, rfft, irfft

ts = np.arange(0, 5 / fc, 1 / (100 * fc))
square_wave_with_delay  = square_wave_delay(ts, fc)
fft_square_wave_with_delay = rfft(square_wave_with_delay)

# visualize fourier transform
plt.figure(figsize=(10,6))
plt.plot(np.abs(fft_square_wave_with_delay))
plt.title('Fourier Transform of the Square Wave with delay')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()

# Use the N most significant harmonics to reconstruct the voltage
N = 100 # Number of harmonics to consider

# Create a copy of the FFT, keeping only the first N harmonics
fft_filtered = np.copy(fft_square_wave_with_delay)
fft_filtered[N:] = 0  # Zero out all coefficients after the Nth harmonic
fft_filtered[-N:] = 0  # Zero out the negative frequency components symmetrically

# Reconstruct the signal using inverse FFT (only using N harmonics)
fft_square_wave_with_delay_reconstructed = irfft(fft_filtered)

# Plot the original square wave and the reconstructed wave
plt.figure(figsize=(10, 6))
plt.plot(ts, square_wave_with_delay, label="Original Square Wave", color='blue', alpha=0.7)
plt.plot(ts, np.real(fft_square_wave_with_delay_reconstructed), label=f'Reconstructed (up to {N} harmonics)', color='red', linestyle='--')
plt.title(f'Square Wave and Reconstructed Signal (up to {N} Harmonics)')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.legend()
plt.grid(True)
plt.show()
```
